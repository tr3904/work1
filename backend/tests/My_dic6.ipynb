```python
# Creating MySQL server connectivity for databases
def server_connectivity(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host = host_name,
            user = user_name,
            password = user_password
        )
        print("MySQL connection is established successfully")
    except Error as err:
        print(f"Error: '{err}'")
    return connection
# Put our MySQL passcode
pw = "1234"
# Database name
db = "bank_4"
connection = server_connectivity('localhost', 'root', pw)
```

```python
# Creating MySQL_python
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database create successful")
    except Error as err:
        print(f"Error: '{err}'")
create_database_query = "Create database bank_4"
create_database(connection, create_database_query)
```

```python
# Connect to database
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host = host_name,
            user = user_name,
            password = user_password,
            database = db_name)
        print("MySQL connection is established successfully")
    except Error as err:
        print(f"Error: '{err}'")
    return connection
```

```python
# Execute sql queries
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query was successful")
    except Error as err:
        print(f"Error: '{err}'")
```

```python
# Reading Queries
def read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as err:
        print(f"Error: '{err}'")
```

```python
# Step 2: Define validation rules (you can customize based on your needs)
def validate_row(row):
    # Example validation rules
    if pd.isnull(row['Date of Account']):  # Example: Check for missing values in 'column1'
        return False
    if not isinstance(row['Rate of Interest'], int):  # Example: Check if 'column2' is an integer
        return False
    return True
```

```python
# Step 3: Filter valid and invalid data
valid_data = []
invalid_data = []

for index, row in df3.iterrows():
    if validate_row(row):
        valid_data.append(row)
    else:
        invalid_data.append(row)
```

```python
# Step 4: Export invalid data (errors and duplicates)
invalid_df = pd.DataFrame(invalid_data)
invalid_df.to_csv('invalid_data.csv', index=False)
```
